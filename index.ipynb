# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.
  
  
  
Source

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df_river = df[df['chas']==1]
df_river.head()
```

```python
df_river.describe()
```

```python
df_land = df[df['chas']==0]
df_land.describe()
```

```python
new_figure = plt.figure(figsize=(10,4))


ax = new_figure.add_subplot(121)
ax2 = new_figure.add_subplot(122)

ax.hist(df_river['medv'], label = 'Boston Suburbs')
ax2.hist(df_land['medv'], label = 'Boston Suburbs')

ax.set_title('On Charels River')
ax2.set_title('Not on River')

ax.set_xlabel("median home value in $10000")
ax2.set_xlabel('median home value in $10000')
ax.legend()
ax2.legend()

plt.show()
```

```python
new_figure = plt.figure(figsize=(14,6))

# Add a subplot to the figure - a new axes
ax = new_figure.add_subplot(121)

# Add a second subplot to the figure - a new axes
ax2 = new_figure.add_subplot(122)

# Generate a line plot on first axes
ax.scatter(df_river['dis'], df_river['crim'])

# Draw a scatter plot on 2nd axes
ax2.scatter(df_land['dis'], df_land['crim'], color='red')


ax.set_xlabel('Distance to Employment Center')
ax.set_ylabel('Per Capita Crime')
ax.set_title('On River')

ax2.set_xlabel('Distance to Employment Center')
ax2.set_ylabel('Per Capita Crime')
ax2.set_title('Not On River')

plt.show()
```

```python
df_LotsOfTeachers = df[df['ptratio']<=19]
df_FewTeachers = df[df['ptratio']>19]
```

```python
df_LotsOfTeachers.describe()
```

```python
df_FewTeachers.describe()
```

```python
new_figure = plt.figure(figsize=(14,6))


ax = new_figure.add_subplot(121)
ax2 = new_figure.add_subplot(122)

ax.hist(df_LotsOfTeachers['crim'], label = 'Boston Suburbs',bins = 20)
ax2.hist(df_FewTeachers['crim'],label = 'Boston Suburbs', bins = 20)

ax.set_title('Lots of Teachers')
ax2.set_title('Fewer Teachers')

ax.set_xlabel('Crimes Per Capita')
ax2.set_xlabel('Crimes Per Capita')
ax.legend()
ax2.legend()
plt.show()
```

```python
new_figure = plt.figure(figsize=(14,6))

# Add a subplot to the figure - a new axes
ax = new_figure.add_subplot(121)

# Add a second subplot to the figure - a new axes
ax2 = new_figure.add_subplot(122)

# Generate a line plot on first axes
ax.scatter(df_LotsOfTeachers['rm'], df_LotsOfTeachers['lstat'])

# Draw a scatter plot on 2nd axes
ax2.scatter(df_FewTeachers['rm'], df_FewTeachers['lstat'], color='red')


ax.set_xlabel('Number of Rooms')
ax.set_ylabel('Percent lower Status')
ax.set_title('Lots of Teachers')

ax2.set_xlabel('Number of Rooms')
ax2.set_ylabel('Percent Lower Status')
ax2.set_title('Fewer Teachers')

plt.show()
```

```python
df_lowTax = df[df['tax']<=330]
df_lowTax.describe()
```

```python
new_figure = plt.figure(figsize=(14,6))


ax = new_figure.add_subplot(121)
ax2 = new_figure.add_subplot(122)

ax.hist(df_highTax['dis'], label = 'Boston Suburbs',bins = 15)
ax2.hist(df_lowTax['dis'], label = 'Boston Suburbs',bins = 15)

ax.set_title('High Tax Areas')
ax2.set_title('Low Tax Areas')

ax.set_xlabel('Distance to Employment Centers')
ax2.set_xlabel('Distance to Employment Centers')
ax.legend()
ax2.legend()
plt.show()
```

```python
new_figure = plt.figure(figsize=(14,6))

# Add a subplot to the figure - a new axes
ax = new_figure.add_subplot(121)

# Add a second subplot to the figure - a new axes
ax2 = new_figure.add_subplot(122)

# Generate a line plot on first axes
ax.scatter(df_highTax['indus'], df_highTax['nox'])

# Draw a scatter plot on 2nd axes
ax2.scatter(df_lowTax['indus'], df_lowTax['nox'], color='red')


ax.set_xlabel('proportion nonretail business acres')
ax.set_ylabel('nitrogen oxides concentration (parts per 10 million)')
ax.set_title('High Tax Areas')

ax2.set_xlabel('proportion non-retail business acres')
ax2.set_ylabel('nitrogen oxides concentration (parts per 10 million)')
ax2.set_title('Lower Tax Areas')

plt.show()
```
